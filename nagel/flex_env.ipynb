```python
from reward import generate_reward_matrix
import numpy as np
from skimage import transform
import matplotlib.pyplot as plt
```

```python

class FlexEnv:
    def __init__(self, precision):
        self.act_name = ["leisure", "home", "shopping", "work"]
        self.precision = precision

        self.activity_reward = []
        for activity in self.act_name:
            self.activity_reward.append(transform.resize(
                generate_reward_matrix(activity), (12 / self.precision, 24 / self.precision), order=3))

    def plot_reward(self):
        fig = plt.figure(constrained_layout=True, figsize=(30, 30), num=1, clear=True)
        X, Y = np.meshgrid(np.arange(0, 24, self.precision), np.arange(0, 12, self.precision))
        for act in range(1, 5):
            pos = 220 + act
            ax = fig.add_subplot(pos, projection='3d')
            surf = ax.plot_surface(X, Y, self.activity_reward[act - 1], cmap='coolwarm')
            ax.view_init(elev=46, azim=120)
            ax.set_xlabel('start_time', fontsize=30)
            ax.set_ylabel('duration', fontsize=30)
            ax.set_zlabel('reward', fontsize=30)
            ax.set_title(self.act_name[act - 1], fontsize=30, color='k')
            fig.colorbar(surf, shrink=0.5, aspect=5)
        plt.show()
```

```python
fe = FlexEnv(1)
fe.plot_reward()
```
