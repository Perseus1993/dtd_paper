```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def utility_fun(umin,umax,alpha,beta,gamma,t_perf):
    return umin + (umax - umin)/np.power((1 + gamma * np.exp(beta * (alpha - t_perf))),1/gamma)
```

s型utility 函数
![blockchain](fig/utility.png)

```python
# 从配置文件加载参数
param_lines = open('config/parameter.ini').readlines()
for line in param_lines:
    word = line.split(" ")
    activity_name = word[0]
    umin = int(word[1])
    umax = int(word[2])
    alpha = int(word[3])
    beta = float(word[4])
    gamma = int(word[5])
    
    # 使用 np.arange(0, 12, 0.1) 来代表0到12小时，步长为0.1小时
    plt.plot(np.arange(0, 12, 0.1), [utility_fun(umin, umax, alpha, beta, gamma, i) for i in np.arange(0, 12, 0.1)], label=activity_name)

plt.legend()
plt.xlabel('Hours')  # 添加横轴标签
plt.ylabel('Utility')
plt.show()
```
