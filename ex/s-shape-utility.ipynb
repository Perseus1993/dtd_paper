```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def utility_fun(umin,umax,alpha,beta,gamma,t_perf):
    return umin + (umax - umin)/np.power((1 + gamma * np.exp(beta * (alpha - t_perf))),1/gamma)
```

```python
#load parameter from config
param_lines = open('config/parameter.ini').readlines()
for line in param_lines:
    word = line.split(" ")
    activity_name = word[0]
    umin = int(word[1])
    umax = int(word[2])
    alpha = int(word[3])
    beta = float(word[4])
    gamma = int(word[5])
    plt.plot(np.arange(120),[utility_fun(umin,umax,alpha,beta,gamma,i/10) for i in np.arange(120)],label = activity_name)
    print(utility_fun(umin,umax,alpha,beta,gamma,6))
plt.legend()
plt.show()
```
