```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
filepath = r"D:\data\new_york\Citywide_Mobility_Survey_-_Trip_Survey_2019.csv"
df = pd.read_csv(filepath)
```

```python
sns.set(font_scale = 0.5)

plt.figure(figsize=(20,12))
sns.displot(
    data=df.isna().melt(value_name="missing"),
    y="variable",
    hue="missing",
    multiple="fill",
    aspect=1.25,
)
plt.savefig("visualizing.png", dpi=300)
plt.show()
plt.close()
```
